```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['Cholesterol_HDL'])
plt.xlabel('Index')
plt.ylabel('Cholesterol_HDL')
plt.title('Cholesterol_HDL Distribution')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
